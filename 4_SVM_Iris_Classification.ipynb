# Support Vector Machine (SVM) Classification on the Iris Dataset


This notebook demonstrates how to use **Support Vector Machine (SVM)** for classification on the Iris dataset. 
SVM is a powerful linear classifier that aims to find the optimal hyperplane separating different classes by maximizing 
the margin between classes. SVMs are effective in high-dimensional spaces and are widely used for classification tasks.


```python

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Load Dataset and Split Data

```python

# Load the Iris dataset
data = load_iris()
X, y = data.data, data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Train the Support Vector Machine (SVM) Model

```python

# Initialize the Support Vector Classifier (SVC) with a linear kernel
svm = SVC(kernel='linear', random_state=42)

# Train the model
svm.fit(X_train, y_train)
```

## Make Predictions and Evaluate the Model

```python

# Make predictions on the test set
y_pred = svm.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"SVM Accuracy: {accuracy:.2f}\n")

# Print the classification report
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=data.target_names))
```

## Confusion Matrix

```python

# Confusion Matrix for visualization
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - SVM on Iris Dataset")
plt.show()
```


### Summary
In this notebook, we used a **Support Vector Machine (SVM)** classifier to classify the Iris dataset. We trained 
the model, evaluated its performance using accuracy and classification report metrics, and visualized 
the results with a confusion matrix. SVM is a robust classifier, especially effective in high-dimensional spaces, 
making it suitable for a wide variety of classification tasks.
