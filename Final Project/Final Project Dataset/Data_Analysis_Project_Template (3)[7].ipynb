### Insight #3

**Insight Description:** Trapezoid of Excellence<br>

The Trapezoid of excellence is a concept developed by sports analysts to identify teams that are the most likely to succeed in the tournamnet. Bracketologists were able to identify that almost every team who left the NCAA tournament a champion, fit this statistical profile. 

The trapezoid considers statistically well rounded teams. Solid offensive and defensive performance, meaning a team is likely to rank in the top 20 or so percent of teams in these areas is a factor in placing a team in the trapezoid. Adjusted tempo (pace) plays a role, as teams who are too fast or too slow likely will not fit the model. 

Defines performance thresholds: 
- Offensive Efficiency: Top 25% → Elite scorers
- Defensive Efficiency: Top 40% (lower value = better defense) → Relaxed to capture more strong teams
- Tempo: Above median → Avoids extremes in playing speed

These values define the boundaries of the “Trapezoid.”

```python
# Calculate top performance thresholds
off_threshold = df_cleaned['Offensive Efficiency'].quantile(0.75) # top 25%
def_threshold = df_cleaned['Defensive Efficiency'].quantile(0.60) # top 40%, adjusted this slightly to include a few more teams whose stats didn't fit perfectly in top 20%
tempo_threshold = df_cleaned['Adjusted Tempo'].quantile(0.50)   # above average tempo
print('Trapezoid of Excellence Thresholds:')
print('Offesnive Efficiency')
print(off_threshold)
print('Defensive Efficiency')
print(def_threshold)
print('Adjusted Tempo')
print(tempo_threshold)
```

```python
# Created a new column based on all criteria
df_cleaned['In Trapezoid'] = (
    (df_cleaned['Offensive Efficiency'] >= off_threshold) &
    (df_cleaned['Defensive Efficiency'] >= def_threshold) &
    (df_cleaned['Adjusted Tempo'] >= tempo_threshold)
)

# who has made the Trapezoid? 
df_cleaned[
    (df_cleaned['In Trapezoid'] == True) & 
    (df_cleaned['Year'] == 2024)
][['Team', 'Year', 'Seed', 'Conference', 'Offensive Efficiency', 'Defensive Efficiency', 'In Trapezoid']]

#note that I had to change my criteria to include more teams, it was too strict and narrow at first only showing two teams
```

Teams said to have been in it in 2024: FL, FAU (on the outside a bit) Gonzaga, Illinois, Duke, Alabama, Michigan State, Houston, Iowa State, and Tenessee. 

```python
# Profiling trapezoid teams to establish a benchmark of statistical excellence
trapezoid_avg = df_cleaned[df_cleaned['In Trapezoid'] == True][
    ['Offensive Efficiency', 'Defensive Efficiency', 'Adjusted Tempo']
].mean()

print(trapezoid_avg)
```

```python
# Testing whether new or hypothetical teams meet the trapezoid criteria
def predict_trapezoid_membership(off_eff, def_eff, tempo):
    return (
        (off_eff >= off_threshold) and
        (def_eff <= def_threshold) and
        (tempo >= tempo_threshold)
    )

# Example: Predict for a hypothetical 2025 team
predict_trapezoid_membership(off_eff=122.5, def_eff=95, tempo=71)
```

### Analysis

The Trapezoid of Excellence was a helpful way to look at which NCAA teams are actually well-rounded — not just good at one thing like scoring a lot. By using offensive efficiency, defensive efficiency, and tempo together, the model helped highlight teams that are strong on both ends of the court and play at a solid pace. What stood out is that the teams that checked all three boxes were often the ones that made deeper runs in the tournament.

One of the more surprising things was realizing that just having a strong offense isn’t always enough. Some teams that were great at scoring didn’t make it far because they either couldn’t stop other teams from scoring or they played way too slowly or too fast to keep control. Teams that balanced all three areas — solid offense, tight defense, and a decent tempo — were the ones that consistently looked like legit contenders. It also became clear how important defense actually is, even if it doesn’t get talked about as much as offense. The teams that quietly had great defensive numbers tended to go far.

Something else that was useful was being able to test and simulate team stats. It gave me a way to see how the model might work for future teams, not just ones from past tournaments. That ability to predict or evaluate a team before the bracket even comes out makes the model feel more powerful and practical — it’s not just about looking back, but also about looking ahead.

In the end, this whole part of the project showed that winning in March Madness isn’t about being amazing at one thing — it’s about being pretty great across the board. The Trapezoid helped make that clearer and gave me a way to measure it in a consistent way.

### Insight #5 (based on visualization #2)

**Tournament Overachievers vs. Underachievers**

```python
#Biggest overachievers in the tournament compared to biggest underachievers (based on seed)

# Expected: Lower seed -> Higher Round Score
df['Performance Delta'] = df['Round Score'] - (9 - df['Seed'])  # 9-seed would be average baseline
# Overachievers: Low seed, high performance
overachievers = df[df['Performance Delta'] > 3].sort_values(by='Performance Delta', ascending=False)
# Underachievers: High seed, poor performance
underachievers = df[df['Performance Delta'] < -2].sort_values(by='Performance Delta')
print("Big Overachievers:\n")
print(overachievers[['Team', 'Seed', 'Elimination Round', 'Year', 'Performance Delta']].reset_index(drop=True))
print("\nBig Underachievers:\n")
print(underachievers[['Team', 'Seed', 'Elimination Round', 'Year', 'Performance Delta']].reset_index(drop=True))
```

A Performance Delta of 0 implies that the team finished roughly where expected based on their seed. Positive values indicate teams that exceeded expectations (overachievers), while negative values indicate those that fell short (underachievers).


Next, we visualized the top overachievers and underachievers:

```python
# Combine top 10 overachievers and underachievers
over_vs_under = pd.concat([
   overachievers[['Team', 'Year', 'Performance Delta']],
   underachievers[['Team', 'Year', 'Performance Delta']]
])
top_deltas = over_vs_under.sort_values(by='Performance Delta', ascending=False).head(10)
bottom_deltas = over_vs_under.sort_values(by='Performance Delta').head(10)
combined = pd.concat([top_deltas, bottom_deltas])
# Label for y-axis
combined['Label'] = combined['Team'] + ' (' + combined['Year'].astype(str) + ')'
# Categorize for coloring
combined['Category'] = combined['Performance Delta'].apply(lambda x: 'Overachiever' if x > 0 else 'Underachiever')
# Sort for visual order
combined_sorted = combined.sort_values(by='Performance Delta')
# Plot with hue assigned
plt.figure(figsize=(10, 6))
sns.barplot(
   data=combined_sorted,
   x='Performance Delta',
   y='Label',
   hue='Category',
   dodge=False,
   palette={'Overachiever': 'seagreen', 'Underachiever': 'indianred'}
)
plt.axvline(0, color='gray', linestyle='--')
plt.title('Top Overachievers and Underachievers in March Madness (2013–2025)')
plt.xlabel('Performance Delta (Actual - Expected Round)')
```

### Analysis

This chart provides a direct visual continuation of the boxplot, identifying and labeling the most dramatic deviations from expected seed performance (with a performance delta of 0 being the baseline). Saint Peter’s (15-seed) and UMBC (16-seed) are standouts among overachievers, while multiple 1-seeds are underachievers such as Virginia and Purdue.

### Insight #6 (based on visualization #2)

**Do strong teams always make it?**  <br>

Finally, we explored whether team strength (as measured by BARTHAG) aligns with tournament advancement. We compared the team strength of champions versus the strongest team that didn’t make the tournament each year. The following code was used to generate this comparison:

```python
# First, reuse earlier logic for missed tournament teams
no_tourney_teams = df[df['Seed'].isnull()]
strongest_missed = no_tourney_teams.loc[
   no_tourney_teams.groupby('Year')['Team Strength'].idxmax()
][['Year', 'Team', 'Team Strength']].rename(columns={'Team': 'Missed Team', 'Team Strength': 'Missed Strength'})


# Get teams who won the tournament
champions = df[df['Elimination Round'] == 'Champions']
champions = champions[['Year', 'Team', 'Team Strength']].rename(columns={'Team': 'Champion Team', 'Team Strength': 'Champion Strength'})


# Merge the two dataframes on Year
comparison_df = pd.merge(strongest_missed, champions, on='Year')


# Calculate the strength gap
comparison_df['Strength Gap'] = comparison_df['Champion Strength'] - comparison_df['Missed Strength']


# Display the comparison
print("Champions vs. Strongest Teams That Missed the Tournament (2013–2025):\n")
print(comparison_df[['Year', 'Champion Team', 'Champion Strength', 'Missed Team', 'Missed Strength', 'Strength Gap']].reset_index(drop=True))
```

This code performs a year-by-year comparison between two specific groups:
- The team that won the tournament (Champion Team)
- The strongest team (by BARTHAG rating) that didn’t make the tournament (Missed Team)

By identifying the top non-qualifier and comparing it with the champion of the same year, this analysis highlights the gap between a team's statistical strength and its actual postseason outcome. The Strength Gap column provides a numeric difference that quantifies how much stronger the champion was compared to the strongest omitted team. In all observed years, the champions had higher BARTHAG values, reinforcing the competitive edge elite teams bring to March Madness.

We then visualized the comparison using this annotated bar chart code:


```python
# Sort by year for consistent x-axis
comparison_df_sorted = comparison_df.sort_values(by='Year')
# Set plot size
plt.figure(figsize=(12, 6))
# Plot both strengths side-by-side
bar_width = 0.4
x = range(len(comparison_df_sorted))
plt.bar([i - bar_width/2 for i in x], comparison_df_sorted['Champion Strength'],
       width=bar_width, label='Champion Team', color='steelblue')
plt.bar([i + bar_width/2 for i in x], comparison_df_sorted['Missed Strength'],
       width=bar_width, label='Missed Strongest Team', color='darkorange')
# Labeling
plt.xticks(x, comparison_df_sorted['Year'], rotation=45)
plt.xlabel('Year')
plt.ylabel('Team Strength (BARTHAG)')
plt.title('Champion Teams vs. Strongest Missed Teams (2013–2025)')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Analysis

This chart visually confirms the numeric analysis.
```python
# Manually build the Gonzaga dataset based on the seasons you had
gonzaga_data = {
   'Year': [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021, 2022, 2023, 2024],
   'Wins': [31, 28, 34, 27, 37, 32, 33, 26, 28, 31, 25],
   'Turnover Rate': [17.2, 17.4, 16.1, 17.1, 16.2, 16.1, 14.9, 16.1, 15.7, 14.8, 14.0],
   'Steal %': [20.8, 17.4, 17.1, 15.1, 17.1, 17.4, 19.0, 20.3, 16.3, 17.8, 16.5]
}


gonzaga_df = pd.DataFrame(gonzaga_data)
```

```python
# Calculate correlation matrix
correlation_matrix = gonzaga_df[['Turnover Rate', 'Steal %', 'Wins']].corr()
print("Correlation Matrix:")
print(correlation_matrix)


# Specific correlations
tor_wins_corr = correlation_matrix.loc['Turnover Rate', 'Wins']
steal_wins_corr = correlation_matrix.loc['Steal %', 'Wins']


print("\nInterpretation:")
print(f"Correlation between Turnover Rate and Wins: {tor_wins_corr:.2f}")
print(f"Correlation between Steal Percentage and Wins: {steal_wins_corr:.2f}")


if abs(tor_wins_corr) > abs(steal_wins_corr):
   print("\nTurn Over Rate has a stronger relationship with Wins than Steal Percentage.")
else:
   print("\nSteal Percentage has a stronger relationship with Wins than Turn Over Rate.")
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Gonzaga Possession Metrics and Wins")
plt.tight_layout()
plt.show()
```

The closer a value is to 1.00 or -1.00, the stronger the relationship. Values near 0 mean almost no relationship:


   Turnover Rate has a very weak positive correlation with Wins (+0.07).


   Steal Percentage has a slightly stronger, but still very weak, positive correlation with Wins (+0.11).


The weak values shown in the blue shades indicate that neither Turnover Rate nor Steal Percentage meaningfully predicts how many games Gonzaga wins.


The heatmap visually confirms that possession metrics alone are not major drivers of success for Gonzaga; their wins are influenced by a broader combination of factors beyond just turnovers and steals.
