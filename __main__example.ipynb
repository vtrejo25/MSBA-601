When Python runs a script, it automatically sets a special built-in variable called __name__. This variable can take one of two values:

1. When the script runs directly (as the main program):

__name__ is set to "__main__".
The code inside if __name__ == '__main__': executes.

2. When the script is imported as a module in another script:

__name__ is set to the module's filename (without .py).
The code inside if __name__ == '__main__': does NOT execute.
