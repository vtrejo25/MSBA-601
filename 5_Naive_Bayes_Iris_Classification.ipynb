# Naive Bayes Classification on the Iris Dataset


This notebook demonstrates how to use **Naive Bayes** for classification on the Iris dataset. 
Naive Bayes is a probabilistic classifier based on Bayes' theorem, assuming that features are 
independent given the class label. Despite this assumption, it performs well on many classification 
tasks, especially with smaller datasets or when quick classification is needed.


```python

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Load Dataset and Split Data

```python

# Load the Iris dataset
data = load_iris()
X, y = data.data, data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Train the Naive Bayes Model

```python

# Initialize the Naive Bayes classifier
nb = GaussianNB()

# Train the model
nb.fit(X_train, y_train)
```

## Make Predictions and Evaluate the Model

```python

# Make predictions on the test set
y_pred = nb.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Naive Bayes Accuracy: {accuracy:.2f}\n")

# Print the classification report
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=data.target_names))
```

## Confusion Matrix

```python

# Confusion Matrix for visualization
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Naive Bayes on Iris Dataset")
plt.show()
```


### Summary
In this notebook, we used a **Naive Bayes** classifier to classify the Iris dataset. We trained 
the model, evaluated its performance using accuracy and classification report metrics, and visualized 
the results with a confusion matrix. Naive Bayes is a simple yet effective model, especially for 
classification tasks with the assumption of independent features, making it a powerful tool for 
quick and efficient classification.
